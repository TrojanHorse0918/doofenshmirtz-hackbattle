```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import random
```

```python
df=pd.read_csv('crime.csv')
```

```python
df 
```

```python
df.dropna()
```

```python
encoder = OneHotEncoder()


encoded_data = encoder.fit_transform(df[['NEIGHBOURHOOD']])


df_encoded = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['NEIGHBOURHOOD']))

df_encoded
```

```python
encoder = OneHotEncoder()


encoded_data = encoder.fit_transform(df[['NEIGHBOURHOOD']])


df_encoded = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['NEIGHBOURHOOD']))


df_encoded
```

```python
scaler = MinMaxScaler()

df_normalize = df[['MONTH','DAY','HOUR','MINUTE','Latitude','Longitude']]
```

```python
df_normalized=scaler.fit_transform(df_normalize)
```

```python
df_normalized_dataframe = pd.DataFrame(df_normalized,columns=df_normalize.columns)
```

```python
df_normalized_dataframe
```

```python
final_df= pd.concat([df_encoded,df_normalized_dataframe],axis=1)
```

```python
final_df
```

```python
final_df=final_df.dropna()
X=final_df[['HOUR','Latitude','Longitude','NEIGHBOURHOOD_Arbutus Ridge']]
num_rows = 476290

random_list = [random.randint(0.0, 1.0) for i in range(num_rows)]
Y = pd.DataFrame({'Crime': random_list})
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

model=Sequential()
```

```python
model.add(Flatten(input_shape=(3,)))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='softmax')) 
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
y=model.predict(x_test)
```

```python
y=pd.DataFrame(y)
count =0
for num in y[0]:
    if num==0:
        count=count+1
        
count
```
